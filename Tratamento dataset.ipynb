```python
import pandas as pd
import random
import numpy as np
import pandasql as ps

pd.options.display.max_rows = 999
```

```python
path = r'D:\PI3\MarketBasketAnalysis\Dataset'

df = pd.read_excel(path+r'\online_retail_II.xlsx',engine='openpyxl')
df.dropna(subset=['Customer ID'],inplace=True)
df
```

```python
df.Country.value_counts()
```

```python
countries = df['Country'].unique()
cod = list(range(len(countries)))
dicto = dict(zip(countries,cod))
df['Group ID'] = df['Country'].map(dicto)
df.sort_values(by='Group ID')
```

```python
len(df.loc[df['Country'] == 'United Kingdom']['Customer ID'].astype(np.int64).unique())
```

```python
grupo00 = list(range(300))
grupo40 = list(range(300,600))
grupo41 = list(range(600,1200))
grupo42 = list(range(1200,1700))
grupo43 = list(range(1700,2335))
grupo44 = list(range(2335,2535))
grupo45 = list(range(2535,2788))
grupo46 = list(range(2788,3088))
grupo47 = list(range(3088,3493))
grupo48 = list(range(3493,3852))
grupo49 = list(range(3852,4035))
```

```python
df2 = pd.DataFrame(df.loc[df['Country'] == 'United Kingdom']['Customer ID'].astype(np.int64).unique().copy(),columns=['Customer ID'])
df2.loc[df2.index.isin(grupo00),'Group ID'] = 0
df2.loc[df2.index.isin(grupo40),'Group ID'] = 40
df2.loc[df2.index.isin(grupo41),'Group ID'] = 41
df2.loc[df2.index.isin(grupo42),'Group ID'] = 42
df2.loc[df2.index.isin(grupo43),'Group ID'] = 43
df2.loc[df2.index.isin(grupo44),'Group ID'] = 44
df2.loc[df2.index.isin(grupo45),'Group ID'] = 45
df2.loc[df2.index.isin(grupo46),'Group ID'] = 46
df2.loc[df2.index.isin(grupo47),'Group ID'] = 47
df2.loc[df2.index.isin(grupo48),'Group ID'] = 48
df2.loc[df2.index.isin(grupo49),'Group ID'] = 49
df2['Group ID'] = df2['Group ID'].astype(np.int64)
df2
```

```python
query = """
select
    df.Invoice as cd_compra,
    df.StockCode as cd_item,
    df.Description as tx_item,
    df.Quantity as nm_quantidade,
    strftime('%Y-%m-%d',df.InvoiceDate) as dt_compra,
    df.Price as nm_vl_item,
    df.Price*df.Quantity as nm_vl_total,
    cast(df."Customer ID" as integer) as cd_cliente,
    case
        when df2."Group ID" is not null then cast(df2."Group ID" as integer)
        else df."Group ID"
    end as cd_grupo_cliente
from
    df
left join
    df2
on
    df."Customer ID" = df2."Customer ID"
order by
    cd_grupo_cliente
"""

df3 = ps.sqldf(query)
```

```python
df3.info()
```

```python
clientes = pd.DataFrame(df3['cd_cliente'].unique(),columns=['cd_cliente'])
clientes.sort_values(by='cd_cliente',inplace=True)
clientes['id_cliente'] = clientes.index 
clientes = clientes[['id_cliente','cd_cliente']]
nomes = pd.read_csv(path+r'\name_gender_dataset.csv',sep=',')
nomes.drop(['Count','Probability'],axis=1,inplace=True)
nomes = nomes.iloc[:len(clientes)]
clientes[['cliente_nome','cliente_sexo']] = nomes[['Name','Gender']]
clientes.reset_index(drop=True,inplace=True)
clientes.to_csv(path+r'\clientes.csv',sep=';',index=False)
clientes
```

```python
cidades = pd.read_csv(path+r'\gb.csv')
query = """
select
    country as tx_pais, iso2 as cd_pais,
    admin_name as cd_estado, sum(population) as nm_qtd_populacao
from
    cidades
group by
    country, iso2,admin_name
order by
    population desc
limit
    47
"""
cidades = ps.sqldf(query)
cidades
```

```python
grupos = pd.DataFrame(df3['cd_grupo_cliente'].unique(),columns=['cd_grupo_cliente'])
grupos['id_grupo'] = grupos.index
grupos[['cd_pais','tx_pais','cd_estado']] = cidades[['cd_pais','tx_pais','cd_estado']]
grupos = grupos[['id_grupo','cd_grupo_cliente','cd_pais','tx_pais','cd_estado']]
grupos.to_csv(path+r'\grupo_cliente.csv',sep=';',index=False)
grupos
```

```python
itens = df3[['cd_item','tx_item']].copy()
itens.sort_values(by='cd_item',inplace=True)
itens.drop_duplicates(subset=['cd_item'],inplace=True)
itens.reset_index(drop=True,inplace=True)
itens['id_item'] = itens.index
itens = itens[['id_item','cd_item','tx_item']]
itens.to_csv(path+r'\itens.csv',sep=';',index=False)
itens
```

```python
df3['id_compra'] = df.index
```

```python
query_df = """
select
    df3.id_compra,
    df3.cd_compra,
    itens.id_item,
    clientes.id_cliente,
    grupos.id_grupo,
    df3.dt_compra,
    df3.nm_quantidade,
    df3.nm_vl_item,
    df3.nm_vl_total
from
    df3
left join
    clientes 
on
    df3.cd_cliente = clientes.cd_cliente
left join
    grupos
on
    df3.cd_grupo_cliente = grupos.cd_grupo_cliente
left join
    itens
on
    df3.cd_item = itens.cd_item
"""
df4 = ps.sqldf(query_df)
df4['nm_vl_item'] = df4['nm_vl_item'].round(2)
df4['nm_vl_total'] = df4['nm_vl_total'].round(2)
df4
```
